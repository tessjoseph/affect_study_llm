```python
import glob
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
from scipy.stats import pearsonr
from sklearn import datasets
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from scipy.stats import kendalltau
from scipy.stats import ttest_ind
import pingouin as pg
```

```python
# path to the directory containing the CSV files for european american participants
path = f'results/final/gpt3.5/us_affect_study_final_hkcgpt3.5*'

# list of all CSV files
files = glob.glob(path)

# read and concatenate all CSV files into a single DataFrame
df_list_hkc = [pd.read_csv(file) for file in files]

df = pd.concat(df_list_hkc, ignore_index=True)

df['group'] = 'Hong Kong Chinese'
```

```python
# path to the directory containing the CSV files for european american participants
path = f'results/final/gpt3.5/us_affect_study_final_aagpt3.5*'

# list of all CSV files
files = glob.glob(path)

# read and concatenate all CSV files into a single DataFrame
df_list_aa = [pd.read_csv(file) for file in files]

df_aa = pd.concat(df_list_aa, ignore_index=True)

df_aa['group'] = 'Asian American'


df = pd.concat([df_aa, df], axis=0)

#drops unnecessary column that was added when combining the dataframe
df = df.drop(columns=['Unnamed: 0'], inplace=False)
```

```python
# path to the directory containing the CSV files for european american participants
path = f'results/final/gpt3.5/us_affect_study_final_usgpt3.5*'

# list of all CSV files
files = glob.glob(path)

# read and concatenate all CSV files into a single DataFrame
df_list_aa = [pd.read_csv(file) for file in files]

df_ea = pd.concat(df_list_aa, ignore_index=True)

df_ea['group'] = 'European American'

combined_df = pd.concat([df_ea, df], axis=0)

#drops unnecessary column that was added when combining the dataframe
combined_df = combined_df.drop(columns=['Unnamed: 0'], inplace=False)
```
